# **Text Pre-Processing**

Text preprocessing adalah suatu proses untuk mengolah data dengan tujuan melakukan seleksi text agar menjadi lebih terstruktur. Serangkaian tahapan perlu dilakukan seperti case folding, tokenizing, filtering dan stemming. Tidak ada aturan yang pasti mengenai urutan dan jenis rangkaian preprocessing, hal ini menyesuaikan dengan kondisi data yang kita miliki serta output yang dibutuhkan.

```python
#Import library yang dibutuhkan untuk tahapan Pre-Processing

import nltk

nltk.download('punkt')

nltk.download('stopwords')

nltk.download('wordnet')

import re #Untuk bekerja dengan regular expression
```

**Menghapus URL dari kolom konten**

```python
df['content'] = df['content'].str.replace('https\S+', ' ', case=False)
df
```

**Merubah Teks Jadi Lower Case**

```python
df['content'] = df['content'].str.lower()
df
```

```python
# substring = '#'
# df[df.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]
```

**Menghapus Mention**

```python
df['content'] = df['content'].str.replace('@\S+', ' ', case=False)
```

**Menghapus Hashtag**

```python
df['content'] = df['content'].str.replace('#\S+', ' ', case=False)
```

**menghapus next karakter**

```python
df['content'] = df['content'].str.replace("\'\w+", ' ', case=False)
```

**Menghapus tanda baca**

```python
df['content'] = df['content'].str.replace("[^\w\s]", ' ', case=False)
```

**Menghapus extra whitespace**

```python
df['content'] = df['content'].str.replace("\s(2)", ' ', case=False)
```

**Mengapus Emoticon**

```python
df['content'] = df['content'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```

**Menghapus komentar yg duplikat**

```python
df.drop_duplicates(subset ="content", keep = 'first', inplace = True)
```

**Filtering (Stopword Removal)**

Filtering adalah tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).

Stopword adalah kata umum yang biasanya muncul dalam jumlah besar dan dianggap tidak memiliki makna. Contoh stopword dalam bahasa Indonesia adalah “yang”, “dan”, “di”, “dari”, dll. Makna di balik penggunaan stopword yaitu dengan menghapus kata-kata yang memiliki informasi rendah dari sebuah teks, kita dapat fokus pada kata-kata penting sebagai gantinya.

```python
#Download dulu library untuk stopwords

import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

# Buatlah list stopwords bahasa Indonesia
stopwords = nltk.corpus.stopwords.words("indonesian")

# Perpanjang list dengan stopword kustom Anda sendiri
my_stopwords = ['gojek','grab']
stopwords.extend(my_stopwords)
```

```python
# Remove stopwords
df['content_token'] = df['content_token'].apply(lambda x: [item for item in x if item not in stopwords])
df.head(10)
```

**Stemming sastrawi**

Sastrawi Python adalah library python sederhana yang memungkinkan Anda untuk mereduksi kata-kata infleksi dalam Bahasa Indonesia (Bahasa Indonesia) ke bentuk dasarnya (stem).

Stemming adalah teknik pemrosesan bahasa alami yang menurunkan infleksi kata ke bentuk akarnya, sehingga membantu dalam pra-pemrosesan teks, kata, dan dokumen untuk normalisasi teks.

```python
#Install terlebih dahulu Sastrawi

!pip install Sastrawi
```

```python
# import Sastrawi package
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
df['stemmed'] = df['content_token'].apply(lambda x: [stemmer.stem(y) for y in x]) # Stem every word.
```

```python
df.head(10)
```

**Menghapus token yang kurang dari 4 karakter**

Kata yang terdiri dari 3 huruf atau kurang, pada data ini, banyak yang tidak memiliki makna, seperti iyh, gk, dl dll, sehingga perlu dihilangkan. Hasilnya disimpan dalam kolom text_string.

```python
df['text_string'] = df['stemmed'].apply(lambda x: ' '.join([item for item in x if len(item)>3]))
```

Kolom text_string akan kita analisis, sehingga perlu diubah menjadi string.

Menambahkan kolom sentimen dengan kriteria, score 1-2 --> negatif dengan kode -1, score 3 adalah netral dengan kode angka 0 dan score 4-5 adalah positif dengan kode 1

**Penerapan TF-IDF**

Term frequency-inverse document frequency adalah text vectorizer yang mengubah teks menjadi vektor yang dapat digunakan. Ini menggabungkan 2 konsep, Term Frequency (TF) dan Document Frequency (DF).

Frekuensi istilah adalah jumlah kemunculan istilah tertentu dalam dokumen. Frekuensi istilah menunjukkan seberapa penting istilah tertentu dalam sebuah dokumen. Frekuensi istilah mewakili setiap teks dari data sebagai matriks yang barisnya adalah jumlah dokumen dan kolomnya adalah jumlah istilah yang berbeda di seluruh dokumen.

Frekuensi dokumen adalah jumlah dokumen yang mengandung istilah tertentu. Frekuensi dokumen menunjukkan seberapa umum istilah tersebut.

Inverse document frequency (IDF) adalah bobot suatu term, hal ini bertujuan untuk mengurangi bobot suatu term jika kemunculan term tersebar di seluruh dokumen. IDF dapat dihitung sebagai berikut: rumus idf.webp

Dimana idfᵢ adalah skor IDF untuk term i, dfᵢ adalah jumlah dokumen yang mengandung term i, dan n adalah jumlah total dokumen. Semakin tinggi DF suatu term, semakin rendah IDF untuk term tersebut. Ketika jumlah DF sama dengan n yang berarti istilah tersebut muncul di semua dokumen, IDF akan menjadi nol, karena log(1) adalah nol, jika ragu masukkan saja istilah ini ke dalam daftar stopword karena tidak menyediakan banyak informasi.

```python
tfid = TfidfVectorizer()
X_final =  tfid.fit_transform(X)
```

**Handling Inbalance**

Dalam Machine Learning dan Data Science sering kita jumpai istilah yang disebut Distribusi Data Tidak Seimbang, umumnya terjadi ketika pengamatan di salah satu kelas jauh lebih tinggi atau lebih rendah dari kelas lainnya. Karena algoritma Machine Learning cenderung meningkatkan akurasi dengan mengurangi kesalahan, mereka tidak mempertimbangkan distribusi kelas.

Masalah ini lazim dalam contoh-contoh seperti Deteksi Penipuan, Deteksi Anomali, Pengenalan wajah, dll.

Teknik ML standar seperti Decision Tree dan Logistic Regression memiliki bias terhadap kelas mayoritas, dan cenderung mengabaikan kelas minoritas. Mereka cenderung hanya memprediksi kelas mayoritas, oleh karena itu, memiliki kesalahan klasifikasi yang besar dari kelas minoritas dibandingkan dengan kelas mayoritas.

Dengan kata yang lebih teknis, jika kita memiliki distribusi data yang tidak seimbang dalam kumpulan data kita, maka model kita menjadi lebih rentan terhadap kasus ketika kelas minoritas memiliki ingatan yang dapat diabaikan atau sangat rendah.



```python
# Handling imbalanced using SMOTE
# smote = SMOTE()
# x_sm,y_sm = smote.fit_resample(X_final,y)

smote = SMOTE(k_neighbors = 1)
x_sm, y_sm = smote.fit_resample(X_final, y)
```
